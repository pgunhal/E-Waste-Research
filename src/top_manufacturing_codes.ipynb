```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

# === CONFIGURATION ===
base_dir = "../data/manufacturing"
country_code_file = "../ref/country_codes_V202501.csv"
hs_name_file = "../data_baci/product_codes_HS12_V202501.csv"  # Assumed format: hs_code, hs_name
years = [2020]
TOP_N_COUNTRIES = 20
TOP_HS_CODES = 10

# === LOAD COUNTRY CODES & HS NAMES ===
country_df = pd.read_csv(country_code_file)
country_lookup = dict(zip(country_df["country_code"], country_df["country_name"]))

hs_name_df = pd.read_csv(hs_name_file)
hs_name_lookup = dict(zip(hs_name_df["code"].astype(str), hs_name_df["description"]))

# === INIT DATA STRUCTURES ===
export_value_by_hs = defaultdict(lambda: defaultdict(float))  # exporter -> hs_code -> value

# === PROCESS TRADE DATA ===
for category in os.listdir(base_dir):
    category_dir = os.path.join(base_dir, category)
    trade_path = os.path.join(category_dir, "trade_data.csv")
    if not os.path.exists(trade_path):
        continue

    df = pd.read_csv(trade_path)
    df = df.rename(columns={"i": "exporter", "k": "hs_code", "v": "value"})
    df["hs_code"] = df["hs_code"].astype(str)
    df = df[df["year"].isin(years)]

    grouped = df.groupby(["exporter", "hs_code"])["value"].sum().reset_index()
    for _, row in grouped.iterrows():
        export_value_by_hs[int(row["exporter"])][row["hs_code"]] += row["value"]

# === CREATE DATAFRAME ===
value_df = pd.DataFrame(export_value_by_hs).T.fillna(0)
value_df.index = value_df.index.map(lambda x: country_lookup.get(x, f"Code {x}"))

# === Filter Top Countries and HS Codes ===
total_exports = value_df.sum(axis=1)
top_exporters = total_exports.nlargest(TOP_N_COUNTRIES).index
value_df_top = value_df.loc[top_exporters]

top_hs_codes = value_df_top.sum().nlargest(TOP_HS_CODES).index.tolist()

# Group rest into "OTHER"
value_df_top["OTHER"] = value_df_top.drop(columns=top_hs_codes, errors="ignore").sum(axis=1)
value_df_top = value_df_top[top_hs_codes + ["OTHER"]]

# === Rename columns to HS Names ===
hs_name_labels = [
    hs_name_lookup.get(code, f"HS {code}") if code != "OTHER" else "OTHER"
    for code in value_df_top.columns
]
value_df_top.columns = hs_name_labels

# === Color Mapping ===
color_palette = plt.get_cmap("tab20").colors
color_map = {name: color_palette[i % len(color_palette)] for i, name in enumerate(hs_name_labels)}
color_map["OTHER"] = "#999999"
bar_colors = [color_map[col] for col in value_df_top.columns]

# === PLOT ===
fig, ax = plt.subplots(figsize=(16, 7))
value_df_top.plot(kind="bar", stacked=True, ax=ax, edgecolor="black", color=bar_colors)

ax.set_title("Top 20 Exporting Countries by Export Value (Split by HS Name + Other)")
ax.set_ylabel("Export Value (thousands USD)")
ax.set_xlabel("Country")
plt.xticks(rotation=45)
ax.legend(title="HS Code Description", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
